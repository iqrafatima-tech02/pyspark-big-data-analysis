```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ECommerce Product Analysis") \
    .getOrCreate()

print("Spark is working")
```

```python
products_df = spark.read.csv(
    r"C:\Users\anam fatima\Desktop\Big Data Project\olist_products_dataset.csv",
    header=True,
    inferSchema=True
)

products_df.show(10)
```

```python
items_df = spark.read.csv(
    r"C:\Users\anam fatima\Desktop\Big Data Project\olist_order_items_dataset.csv",
    header=True,
    inferSchema=True
)

items_df.show(10)
```

```python
joined_df = products_df.join(items_df, on="product_id", how="inner")

most_expensive = joined_df.orderBy(joined_df.price.desc())

most_expensive.select("product_id", "product_category_name", "price").show(1)
```

```python
category_group = products_df.groupBy("product_category_name").count()

category_group.show()
```

```python
product_count = products_df.groupBy("product_category_name").count()

product_count.orderBy("count", ascending=False).show()
```
